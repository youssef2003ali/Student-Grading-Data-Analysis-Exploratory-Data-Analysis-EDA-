# Data Cleaning & Procecssing 
Filling missing values & Dropping unwanted Features!
Here three columns have missing values which are : Attendance , Assignments_AVg and Parent_Education_Level
Removing unnecessary columns such as : Student_ID,First_Name,Last_Name,Email

```python
print("Filling Null Values")
dataset['Attendance (%)'] = dataset['Attendance (%)'].fillna(dataset['Attendance (%)'].mean())
dataset['Assignments_Avg'] = dataset['Assignments_Avg'].fillna(dataset['Assignments_Avg'].mean())
dataset['Parent_Education_Level'] = dataset['Parent_Education_Level'].fillna(dataset['Parent_Education_Level'].mode()[0])
```

```python
dataset.drop(columns=['Student_ID', 'First_Name', 'Last_Name', 'Email'], inplace=True)
```

# Cheecking for OUtliers using IQR
The purpose of this section is to identify and handle any potential outliers in the dataset. Outliers can significantly affect the performance of machine learning models and statistical analyses, so itâ€™s important to detect and address them accordingly.

```python
for col in dataset.select_dtypes(include=['number']).columns:
    Q1 = dataset[col].quantile(0.25)
    Q3 = dataset[col].quantile(0.75)
    IQR = Q3-Q1
    lowerbounds = Q1 - 1.5 * IQR
    upperbounds = Q3 + 1.5 * IQR
    outliers = dataset[(dataset[col] < lowerbounds) | (dataset[col] > upperbounds)]
    print (f"Number of outliers at {col} equals {len(outliers)}")
```

# Data Visualization
This section aims to visualize the data to understand its distribution, detect patterns, and uncover relationships between variables. Data visualization plays a critical role in identifying trends, correlations, and anomalies in the dataset, which can inform further analysis and decision-making.

```python
num_cols = dataset.select_dtypes(include=['number']).columns
colors = sns.color_palette("magma")
plt.figure(figsize=(15, 15))
for i , col in enumerate(num_cols):
    plt.subplot(4,4,i+1)
    sns.histplot(dataset[col],bins = 20,kde=True ,color=colors[i%len(colors)],alpha=0.75)
    plt.title(f"Distrbution of {col}")
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.suptitle("Distrbution of Numerical Features",fontsize=16)

plt.show()
```
